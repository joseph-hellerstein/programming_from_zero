# Python flow of control

* ``if``, ``elif``, ``else``
* ``for``
* ``def`` (functions) and requirements for writing functions

```python
# Determine if a number is prime
number = 7
# Loop to determine if number is divisible by something other than 1 and itself
is_prime = True
for i in range(2, number):
  if number % i == 0:
    is_prime = False
print(is_prime)
```

## From scrip to function

```python
def is_prime(number: int) -> bool:
    # Checks if number is prime
    is_prime = True
    for i in range(2, number):
        if number % i == 0:
            is_prime = False
    return is_prime

# Test
assert(is_prime(7))
assert(not is_prime(8))
print("OK!")
```
