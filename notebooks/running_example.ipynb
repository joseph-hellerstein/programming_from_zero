## ANES PARTISANSHIP VARIABLES

## Core Party Identification Variables

### **VCF0301 - Party Identification (7-Point Scale)**
**Description:** The foundational measure of party identification using the classic Michigan model.

**Scale (1-7):**
1. Strong Democrat
2. Weak Democrat
3. Independent-leaning Democrat
4. Independent (pure independent)
5. Independent-leaning Republican
6. Weak Republican
7. Strong Republican

**Missing codes:** 0, 9 = NA/DK/Refused

**Years available:** 1952-2024

**Key uses:** 
- Primary measure of partisan attachment
- Predicts voting behavior, policy preferences
- Tracks partisan realignments over time

---

### **VCF0302 - Initial Party ID Response**
**Description:** Records respondent's first answer to the party ID question before follow-up probes.

**Scale:**
1. Democrat
2. Republican
3. Independent
4. Other party
5. No preference

**Years available:** 1952-2024

**Key uses:**
- Distinguishes initial response from leaning
- Useful for understanding strength of partisan attachment

---

### **VCF0303 - Party Identification (3-Category Summary)**
**Description:** Simplified party identification that includes leaners with partisans.

**Scale (1-3):**
1. Democrat (includes Strong, Weak, and Independent-leaning Democrats)
2. Independent (pure independents only - no partisan lean)
3. Republican (includes Strong, Weak, and Independent-leaning Republicans)

**Years available:** 1952-2024

**Key uses:**
- Standard breakdown variable for cross-tabulations
- Most commonly used in ANES Guide visualizations
- Treats leaners as "hidden partisans"

---

### **VCF0305 - Strength of Partisanship**
**Description:** Measures intensity of partisan attachment regardless of direction.

**Scale (0-4):**
- 1 = Independent (no partisan leaning)
- 2 = Leaning Independent (either direction)
- 3 = Weak partisan (either party)
- 4 = Strong partisan (either party)
- 0 = No response

**Years available:** 1952-2024

**Key uses:**
- Studies partisan polarization
- Examines intensity vs. direction of partisanship
- Tracks growth/decline of pure independents

---

### **VCF0306 - Father's Party Identification**
**Description:** Respondent's report of their father's party identification.

**Scale:** Same 7-point scale as VCF0301

**Years available:** 1952-2024

**Key uses:**
- Intergenerational transmission of partisanship
- Parental socialization effects

---

### **VCF0307 - Mother's Party Identification**
**Description:** Respondent's report of their mother's party identification.

**Scale:** Same 7-point scale as VCF0301

**Years available:** 1952-2024

**Key uses:**
- Intergenerational transmission of partisanship
- Research shows mothers may have stronger influence than previously thought

---

## Party Affect and Attitudes

### **VCF0218 - Thermometer: Democratic Party**
**Description:** Feeling thermometer rating of the Democratic Party.

**Scale (0-100):**
- 0 = Very cold/unfavorable
- 50 = Neutral
- 100 = Very warm/favorable

**Years available:** 1978-2024

**Key uses:**
- Affective polarization research
- Party attitudes beyond identification
- Measuring "warmth" toward parties

---

### **VCF0224 - Thermometer: Republican Party**
**Description:** Feeling thermometer rating of the Republican Party.

**Scale (0-100):**
- 0 = Very cold/unfavorable
- 50 = Neutral
- 100 = Very warm/favorable

**Years available:** 1978-2024

**Key uses:**
- Affective polarization research
- Party attitudes beyond identification
- Measuring "warmth" toward parties

---

### **VCF0290 - Thermometer Average: Major Parties**
**Description:** Average of Democratic and Republican party thermometer ratings.

**Scale (0-100):** Average of VCF0218 and VCF0224

**Years available:** 1978-2024

**Key uses:**
- Overall attitude toward the party system
- System support measures

---

### **VCF0291 - Thermometer Index: Major Parties**
**Description:** Difference between own-party and opposing-party thermometer ratings.

**Calculation:** 
- For Democrats: Democratic Party thermometer - Republican Party thermometer
- For Republicans: Republican Party thermometer - Democratic Party thermometer

**Scale:** -100 to +100 (higher = stronger partisan affect)

**Years available:** 1978-2024

**Key uses:**
- **Affective polarization** (primary measure)
- Tracks increase in partisan hostility over time
- Used in ~1,200+ citations since 2012

---

### **VCF0314-VCF0321 - Party Likes/Dislikes**
**Description:** Open-ended responses about what respondent likes/dislikes about each party.

**Variables:**
- VCF0314: Democratic Party likes (number of mentions)
- VCF0315: Democratic Party dislikes (number of mentions)
- VCF0316: Democratic Party affect (likes - dislikes)
- VCF0317: Democratic Party salience (likes + dislikes)
- VCF0318: Republican Party likes (number of mentions)
- VCF0319: Republican Party dislikes (number of mentions)
- VCF0320: Republican Party affect (likes - dislikes)
- VCF0321: Republican Party salience (likes + dislikes)

**Scale:** 0-5+ mentions (varies by year)

**Years available:** 1952-2024

**Key uses:**
- Understanding reasons for party preferences
- Party image research
- Measuring ambivalence

---

## Vote Choice Variables (Dependent Variables Often Predicted by Partisanship)

### **VCF0704 - Presidential Vote (2-Party)**
**Description:** Two-party presidential vote choice.

**Scale:**
1. Democrat
2. Republican

**Missing codes:** 0 = Did not vote/other

**Years available:** 1948-2024 (presidential years)

**Key uses:**
- Primary dependent variable in voting behavior research
- Testing party ID's predictive power

---

### **VCF0706 - House Vote (2-Party)**
**Description:** Two-party House vote choice.

**Scale:**
1. Democrat
2. Republican

**Missing codes:** 0 = Did not vote/other

**Years available:** 1948-2024

**Key uses:**
- Congressional voting behavior
- Coattail effects
- Partisan loyalty in down-ballot races

---

## Key Relationships Between Variables

### **Party ID → Vote Choice**
- Strong partisans vote for their party ~90% of the time
- Weak partisans ~75% loyalty
- Leaning independents ~60-70% loyalty (varies over time)

### **Party ID → Ideology**
- Increasing correlation since 1970s ("ideological sorting")
- 2020s: Most Democrats are liberal, most Republicans are conservative
- 1970s: Much more ideological diversity within parties

### **Party ID → Feeling Thermometers**
- Strong correlation between party ID and own-party warmth
- **Affective polarization**: Increasing gap between own-party and other-party ratings
- 1970s: ~30% had "cold" feelings toward opposing party
- 2020s: ~80% have "cold" feelings toward opposing party

## DATE AND TIMING VARIABLES

### **VCF0004 - Study Year**
**Description:** Year of the ANES study.

**Scale:** 
- 1948, 1952, 1954, 1956, 1958, 1960, 1962, 1964, 1966, 1968, 1970, 1972, 1974, 1976, 1978, 1980, 1982, 1984, 1986, 1988, 1990, 1992, 1994, 1996, 1998, 2000, 2002, 2004, 2008, 2012, 2016, 2020, 2024

**Years available:** All years 1948-2024

**Key uses:**
- Identifying which election year the case is from
- Time-series analysis across elections
- Creating year fixed effects in models

---

### **VCF0006 - Case ID**
**Description:** Unique identifier for each respondent within a study year.

**Scale:** 4-digit numeric ID (varies by year, typically 1-9999)

**Years available:** All years 1948-2024

**Key uses:**
- Linking cases across pre/post-election waves
- Merging with other datasets
- Panel data linking (when applicable)

**Note:** Combined with VCF0004 creates unique identifier across all years

---

### **VCF1015 - Days Before Election (Pre-Election Interview)**
**Description:** Number of days before the election that the pre-election interview was conducted.

**Scale:** 
- Positive integers = days before election
- Typically ranges from 1 to 60+ days

**Missing codes:**
- 0 = Inap, no pre-election interview
- Negative values = various missing data codes

**Years available:** Presidential election years with pre-election interviews

**Key uses:**
- Campaign exposure analysis (early vs. late interviews)
- Controlling for timing effects
- Understanding when opinions were measured
- Identifying last-minute deciders vs. early deciders

**Example interpretation:**
- 60 = Interviewed 60 days (2 months) before election
- 7 = Interviewed 1 week before election
- 1 = Interviewed day before election

---

### **VCF1016 - Days After Election (Post-Election Interview)**
**Description:** Number of days after the election that the post-election interview was conducted.

**Scale:** 
- Positive integers = days after election
- Typically ranges from 1 to 90+ days

**Missing codes:**
- 0 = Inap, no post-election interview
- Negative values = various missing data codes

**Years available:** All years with post-election interviews (1948-2024)

**Key uses:**
- Post-election reflection and memory effects
- Controlling for interview timing in turnout validation
- Winner/loser effects over time
- Distinguishing immediate vs. delayed post-election responses

**Example interpretation:**
- 1 = Interviewed day after election
- 7 = Interviewed 1 week after election
- 30 = Interviewed 1 month after election
- 60 = Interviewed 2 months after election

**Research note:** Earlier post-election interviews may be more accurate for turnout reporting but affected by immediate emotional reactions to results.

---

### **VCF0017 - Mode of Interview**
**Description:** Method used to conduct the interview.

**Scale (varies by year, typical codes):**
- 1 = Face-to-face (in person)
- 2 = Telephone
- 3 = Mail
- 4 = Web/Internet
- 5 = Video (2020, 2024)

**Years available:** 1948-2024

**Key uses:**
- Mode effects analysis
- Data quality assessment
- Understanding social desirability bias
- Controlling for measurement differences

**Historical note:** 
- 1948-2012: Primarily face-to-face
- 2012: Introduction of web mode
- 2020: No face-to-face due to COVID-19
- 2024: Return of face-to-face plus web/video

## GEOGRAPHIC VARIABLES (PUBLIC USE)

### **VCF0112 - Census Region**
**Description:** Four-region Census classification of respondent's location.

**Scale (1-4):**
- 1 = Northeast (New England + Mid-Atlantic)
- 2 = Midwest (East North Central + West North Central)
- 3 = South (South Atlantic + East South Central + West South Central)
- 4 = West (Mountain + Pacific)

**Years available:** 1948-2024

**Key uses:**
- Regional voting patterns
- North/South comparisons
- Regional realignment analysis
- Standard demographic control

**Census divisions within regions:**
- **Northeast:** Maine, NH, VT, MA, RI, CT, NY, NJ, PA
- **Midwest:** OH, IN, IL, MI, WI, MN, IA, MO, ND, SD, NE, KS
- **South:** DE, MD, DC, VA, WV, NC, SC, GA, FL, KY, TN, AL, MS, AR, LA, OK, TX
- **West:** MT, ID, WY, CO, NM, AZ, UT, NV, WA, OR, CA, AK, HI

---

### **VCF0113 - Political South/Non-South**
**Description:** Binary classification based on historical political regions.

**Scale (1-2):**
- 1 = South
- 2 = Non-South

**Years available:** 1948-2024

**Key uses:**
- **Southern realignment** research (shift from Democratic to Republican South)
- Civil rights era political change
- Regional political culture
- Most commonly used in ANES Guide breakdowns

**States coded as "South":**
Typically includes the 11 states of the former Confederacy plus border states:
- AL, AR, FL, GA, KY, LA, MS, NC, SC, TN, TX, VA, and sometimes OK, WV

**Research importance:** Critical for understanding:
- Civil Rights Movement's political impact
- "Solid South" Democratic era (pre-1960s)
- Republican Southern Strategy (1960s-1980s)
- Contemporary partisan geography

## IMPORTANT NOTES AND CAVEATS

### **Date/Timing Variables:**
1. **Interview dates vary** - pre-election interviews typically span 6-8 weeks
2. **Post-election timing matters** - memory decays, emotions change
3. **Campaign dynamics** - early interviews miss late campaign events
4. **Mode and timing often confounded** - web interviews may be completed faster

### **Geographic Variables:**
1. **Boundaries change** - congressional districts redistricted every 10 years
2. **Metropolitan definitions evolve** - what counts as "urban" has changed
3. **Regional coding consistent** - but political meaning of "South" has shifted
4. **Privacy protections** - most detailed geography is restricted access only
5. **Sample design** - not all geographic areas sampled every year

### **Missing Data:**
- Alaska and Hawaii: Not included in face-to-face samples
- Some geographic variables have gaps in coverage
- Always check codebook for year-specific availability

### **Weighting:**
- Use appropriate weights (VCF0009x/y/z) when analyzing geographic patterns
- Weights adjust for unequal selection probabilities and non-response
- Different weights for face-to-face vs. web vs. combined samples
